```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
def SIR(Y,s,R0):
    """
    Ecuaciones diferenciales del modelo SEIR
    Inputs:
               Y: 3-dimensional np.array conteniendo variables S,I,R
               s: time float64
               R0: float64
    retorna
        3-dimensional np.array S,I,R.
    """
    dSds=-R0*Y[0]*Y[1]
    dIds=R0*Y[0]*Y[1]-Y[1]
    dRds=Y[1]
    return dSds,dIds,dRds
```

```python
I_0=.001
R_0=0.0
S_0=1-I_0
R0=2.0
t=np.linspace(0,100,10000)
sol=odeint(SIR,[S_0,I_0,R_0],t,args=(R0,))
```

```python
plt.plot(t,sol)
```
